```python
import pandas as pd
import numpy as np
verlander = pd.read_csv('~/Downloads/verlander_data.csv',index_col=0)
scherzer = pd.read_csv('~/Downloads/scherzer_data.csv',index_col=0)
lester =  pd.read_csv('~/Downloads/lester_data.csv',index_col=0)
hamels =  pd.read_csv('~/Downloads/hamels_data.csv',index_col=0)
```

```python
for pitcher in pitchers:
    pitcher.drop(['index'],axis=1,inplace=True)
```

```python
hamels = hamels.loc[:, ~hamels.columns.str.contains('^Unnamed')]
lester = lester.loc[:, ~lester.columns.str.contains('^Unnamed')]
verlander = verlander.loc[:, ~verlander.columns.str.contains('^Unnamed')]
scherzer = scherzer.loc[:, ~scherzer.columns.str.contains('^Unnamed')]
```

```python
hamels_test = hamels.copy()
verlander_test = verlander.copy()
scherzer_test = scherzer.copy()
lester_test = lester.copy()
```

```python
pitchers = [verlander,lester,hamels,scherzer]
test = [verlander_test,lester_test,hamels_test,scherzer_test]
for pitcher in pitchers:
    for index,row in pitcher.iterrows():
        if row["game_year"] in [2017,2018,2019]:
            pitcher.drop(index,inplace=True)
hamels.reset_index(inplace=True)
verlander.reset_index(inplace=True)
scherzer.reset_index(inplace=True)
lester.reset_index(inplace=True)
```

```python
for pitcher in test:
    for index,row in pitcher.iterrows():
        if row["game_year"] not in [2016,2017,2018,2019]:
            pitcher.drop(index,inplace=True)
hamels_test.reset_index(inplace=True)
verlander_test.reset_index(inplace=True)
scherzer_test.reset_index(inplace=True)
lester_test.reset_index(inplace=True)
```

```python
pd.set_option('display.max_columns', 500)
verlander,lester,hamels,scherzer = pitchers
verlander_test,lester_test,hamels_test,scherzer_test = test
```
