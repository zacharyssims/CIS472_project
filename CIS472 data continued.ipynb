```python
import pandas as pd
import numpy as np
verlander = pd.read_csv('~/Downloads/verlander_data.csv')
scherzer = pd.read_csv('~/Downloads/scherzer_data.csv')
lester =  pd.read_csv('~/Downloads/lester_data.csv')
hamels =  pd.read_csv('~/Downloads/hamels_data.csv')
```

```python
for i in range(len(hamels)):
    if hamels['home_team'].iloc[i] == '1' or hamels['home_team'].iloc[i] == 1:
        hamels['home_team'].iloc[i] = 1
    else:
        hamels['home_team'].iloc[i] = 0    
```

```python
for i in range(len(hamels)):
    if hamels['home_team'].iloc[i] == '1' or hamels['home_team'].iloc[i] == 1:
        hamels['home_team'].iloc[i] = 1
    else:
        hamels['home_team'].iloc[i] = 0    
```

```python
pitchers = [verlander,scherzer,lester,hamels]
for pitcher in pitchers:
    for column in pitcher.columns:
        print("column:",column)
        print("nas:",sum(pitcher[column] != pitcher[column] ))
```

```python
for pitcher in pitchers:
    for index,row in pitcher.iterrows():
        if row["release_speed"] != row['release_speed']\
        or row["zone"] != row["zone"]\
        or row["pfx_x"] != row["pfx_x"]\
        or row["pfx_z"] != row["pfx_z"]\
        or row["plate_x"] != row["plate_x"]\
        or row["plate_z"] != row["plate_z"]:
            pitcher.drop(index,inplace=True)
hamels.reset_index(inplace=True)
verlander.reset_index(inplace=True)
scherzer.reset_index(inplace=True)
lester.reset_index(inplace=True)
```

```python
pitchers = [verlander,lester,hamels,scherzer]
for pitcher in pitchers:
    pitcher.drop(['release_pos_x','release_pos_y','release_pos_z'],axis=1,inplace=True)
```

```python
hamels = hamels.loc[:, ~hamels.columns.str.contains('^Unnamed')]
```

```python
lester = lester.loc[:, ~lester.columns.str.contains('^Unnamed')]
verlander = verlander.loc[:, ~verlander.columns.str.contains('^Unnamed')]
scherzer = scherzer.loc[:, ~scherzer.columns.str.contains('^Unnamed')]
```

```python
hamels_test = hamels[np.logical_or(hamels['game_year'] == 2017,np.logical_or(hamels['game_year'] == 2018,hamels['game_year'] == 2019))]
```

```python
verlander_test = verlander[np.logical_or(verlander['game_year'] == 2017,np.logical_or(verlander['game_year'] == 2018,verlander['game_year'] == 2019))]
lester_test = lester[np.logical_or(lester['game_year'] == 2017,np.logical_or(lester['game_year'] == 2018,lester['game_year'] == 2019))]
scherzer_test = scherzer[np.logical_or(scherzer['game_year'] == 2017,np.logical_or(scherzer['game_year'] == 2018,scherzer['game_year'] == 2019))]
```

```python
pitchers = [verlander,lester,hamels,scherzer]
test = [verlander,lester,hamels,scherzer]
for pitcher in pitchers:
    for index,row in pitcher.iterrows():
        if row["game_year"] in [2017,2018,2019]:
            pitcher.drop(index,inplace=True)
hamels.reset_index(inplace=True)
verlander.reset_index(inplace=True)
scherzer.reset_index(inplace=True)
lester.reset_index(inplace=True)
```

```python
for pitcher in test:
    for index,row in pitcher.iterrows():
        if row["game_year"] in [2015,2016,2017,2018,2019]:
            pitcher.drop(index,inplace=True)
hamels.reset_index(inplace=True)
verlander.reset_index(inplace=True)
scherzer.reset_index(inplace=True)
lester.reset_index(inplace=True)
```

```python
pd.set_option('display.max_columns', 500)
verlander,lester,hamels,scherzer = pitchers
```
