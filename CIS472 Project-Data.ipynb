```python
#Lester:
## BOS(2008-2014) OAK(2014) CHC(2015-2019)
for i in range(len(lester)):
    if lester['home_team'].iloc[i] == 'BOS':
        if lester['game_year'].iloc[i] in [2008,2009,2010,2011,2012,2013,2014]:
            lester['home_team'].iloc[i] = 1
        else: lester['home_team'].iloc[i] = 0
    elif lester['home_team'].iloc[i] == 'CHC':
        if lester['game_year'].iloc[i] in [2015,2016,2017,2018,2019]:
            lester['home_team'].iloc[i] = 1
        else:
            lester['home_team'].iloc[i] = 0
    elif lester['home_team'].iloc[i] == 'OAK':
        if lester['away_team'].iloc[i] == 'BOS':
            if lester['game_year'].iloc[i] == 2014:
                lester['home_team'].iloc[i] = 0
            else:
                lester['home_team'].iloc[i] = 0
        elif lester['game_year'].iloc[i] == 2014:
            lester['home_team'].iloc[i] = 1
        else: lester['home_team'].iloc[i] = 0
    else:
        lester['home_team'].iloc[i] = 0
```
